````python
import random
from sympy import isprime, mod_inverse
from math import gcd

def generate_prime():
    while True:
        num = random.randint(400, 500)
        if isprime(num):
            return num

# RSA Key Generation
p = generate_prime()
q = generate_prime()
n = p * q
phi_n = (p - 1) * (q - 1)

# Choosing a secure public key exponent 'e'
e = 65537  # Commonly used prime exponent
if gcd(e, phi_n) != 1:  # If not coprime, regenerate primes
    p = generate_prime()
    q = generate_prime()
    n = p * q
    phi_n = (p - 1) * (q - 1)

# Compute private key 'd'
d = mod_inverse(e, phi_n)

# Display keys
print(f"Public Key (e, n): ({e}, {n})")
print(f"Private Key (d, n): ({d}, {n})")

# Encryption Function
sep = "-"
def encrypt(message, e, n):
    return f"{sep}".join([str(pow(ord(char), e, n)) for char in message])

# Decryption Function
def decrypt(encrypted_text, d, n):
    return ''.join(chr(pow(int(char), d, n)) for char in encrypted_text.split(f"{sep}"))

# Simulating Communication
message = "HELLO"
print(f"\nJohn sends message: {message}")

# Encryption
encrypted_msg = encrypt(message, e, n)
print(f"Encrypted Message: {encrypted_msg}")

# Decryption
decrypted_msg = decrypt(encrypted_msg, d, n)
print(f"Merry Receives (Decrypted): {decrypted_msg}")
````

````python
# Diffie-Hellman Key Exchange Demonstration

# Public Parameters (agreed upon by both parties)
n = 11  # Prime modulus (shared publicly)
g = 13  # Base (shared publicly)

# Private Keys (kept secret by each party)
x = 1554  # Private key of Party A
y = 4112  # Private key of Party B

# Step 1: Generate Public Keys
# Each party computes their public key using the formula (g^private_key) % n
A = (g ** x) % n  # Public key of Party A
B = (g ** y) % n  # Public key of Party B

# Step 2: Compute Shared Secret Keys
# Each party uses the other party's public key and their own private key to compute the shared secret
k_1 = (B ** x) % n  # Shared secret key computed by Party A
k_2 = (A ** y) % n  # Shared secret key computed by Party B

# Results
print("=== Diffie-Hellman Key Exchange ===")
print(f"Public Parameters: n = {n}, g = {g}")
print(f"Private Key of Party A (x): {x}")
print(f"Private Key of Party B (y): {y}")
print(f"Public Key of Party A (A): {A}")
print(f"Public Key of Party B (B): {B}")
print(f"Shared Secret Key computed by Party A (k_1): {k_1}")
print(f"Shared Secret Key computed by Party B (k_2): {k_2}")

# Verification of Key Exchange
if k_1 == k_2:
    print("Key Exchange Successful: Shared Secret Key matches!")
else:
    print("Key Exchange Failed: Shared Secret Key does not match.")
````

````python
import hashlib
import time
import os

def generate_test_messages(base_size, multiplier, count):
    """
    Generate a list of test messages of varying sizes.
    :param base_size: The base size (in bytes) of the smallest message.
    :param multiplier: The factor by which each subsequent message size increases.
    :param count: The number of messages to generate.
    :return: A list of byte strings of varying sizes.
    """
    messages = []
    for i in range(count):
        size = base_size * (multiplier ** i)
        messages.append(os.urandom(size))  # Generate random bytes of the given size
    return messages

def hash_message(algorithm, message):
    """
    Hash a message using the specified algorithm.
    :param algorithm: The hashing algorithm to use (e.g., 'md5', 'sha1').
    :param message: The message to hash.
    :return: The hash digest as a hexadecimal string.
    """
    hasher = hashlib.new(algorithm)
    hasher.update(message)
    return hasher.hexdigest()

def test_integrity_and_performance(messages, algorithms):
    """
    Test the integrity and performance of hash algorithms on varying message sizes.
    :param messages: A list of messages to hash.
    :param algorithms: A list of hash algorithms to test (e.g., ['md5', 'sha1']).
    """
    print("=== Hashing Performance and Integrity Test ===")
    for algorithm in algorithms:
        print(f"\nAlgorithm: {algorithm.upper()}")
        for i, message in enumerate(messages):
            start_time = time.time()
            digest = hash_message(algorithm, message)
            end_time = time.time()
            elapsed_time = end_time - start_time
            print(f"Message Size: {len(message)} bytes | Hash: {digest} | Time: {elapsed_time:.6f} seconds")

def main():
    # Configuration for test message sizes
    base_size = 128         # Base size in bytes
    multiplier = 2          # Factor for size increase
    message_count = 5       # Number of messages to test

    # Generate test messages of varying sizes
    messages = generate_test_messages(base_size, multiplier, message_count)

    # Hashing algorithms to test
    algorithms = ['md5', 'sha1']

    # Test integrity and performance
    test_integrity_and_performance(messages, algorithms)

if __name__ == "__main__":
    main()
````

### **2. Common Nmap Scanning Options and Examples**

Below are the most frequently used Nmap scan types, with example commands. Replace `target` with the IP address or domain you wish to scan.

#### **A. Scan for Open Ports (TCP SYN Scan)**
This is the default and most popular scan, also known as a "half-open" scan.
```bash
nmap -sS target
```
- Scans the 1000 most common TCP ports and reports their state (open, closed, filtered, etc.)[4][5][7].

#### **B. OS Fingerprinting**
Detects the operating system of the target.
```bash
nmap -O target
```
- Attempts to determine the target's OS based on TCP/IP stack responses[8][9].

#### **C. Ping Scan (Host Discovery)**
Checks which hosts are up without scanning ports.
```bash
nmap -sn target
```
- Only pings the target(s) to see if they are online, without port scanning[9].

#### **D. TCP Port Scan**
Scans specified TCP ports.
```bash
nmap -p 22,80,443 target
```
- Scans only ports 22 (SSH), 80 (HTTP), and 443 (HTTPS)[4][7].

#### **E. UDP Port Scan**
Scans for open UDP ports (slower than TCP scans).
```bash
sudo nmap -sU target
```
- Requires root privileges. UDP scans can take longer due to the protocol's nature[7][9].

#### **F. Xmas Scan**
Sends packets with FIN, URG, and PSH flags set ("Xmas tree" scan).
```bash
sudo nmap -sX target
```
- Useful for evading some firewalls and identifying open ports (works mainly on Unix-based systems)[1][7].

---

### **3. Additional Useful Nmap Options**

- **Scan all ports:**  
  ```bash
  nmap -p- target
  ```
  Scans all 65535 TCP ports[4][7].

- **Service/version detection:**  
  ```bash
  nmap -sV target
  ```
  Identifies service and version info on open ports[9].

- **Aggressive scan (combines OS, version, script, and traceroute):**  
  ```bash
  nmap -A target
  ```

### **1. View Existing Rules**

Before making changes, check your current firewall rules:
```bash
sudo iptables -L -v
```
This shows all rules with packet and byte counts[6][5].

---

### **2. Set Default Policies**

For a secure setup, set the default policy to DROP, which blocks all traffic unless explicitly allowed:
```bash
sudo iptables -P INPUT DROP
sudo iptables -P FORWARD DROP
sudo iptables -P OUTPUT ACCEPT
```
- INPUT: Controls incoming packets.
- FORWARD: Controls packets routed through your system.
- OUTPUT: Controls outgoing packets[5][8].

---

### **3. Allow Essential Traffic**

#### **a. Allow Localhost (Loopback)**
```bash
sudo iptables -A INPUT -i lo -j ACCEPT
sudo iptables -A OUTPUT -o lo -j ACCEPT
```
This ensures local applications can communicate[3][7].

#### **b. Allow Established and Related Connections**
```bash
sudo iptables -A INPUT -m conntrack --ctstate ESTABLISHED,RELATED -j ACCEPT
```
This allows responses to your outgoing requests (like web browsing, updates, etc.)[8].

#### **c. Allow SSH (Port 22)**
```bash
sudo iptables -A INPUT -p tcp --dport 22 -j ACCEPT
```
This enables remote management if you use SSH[8][3].

#### **d. Allow HTTP and HTTPS (Ports 80 and 443)**
```bash
sudo iptables -A INPUT -p tcp --dport 80 -j ACCEPT
sudo iptables -A INPUT -p tcp --dport 443 -j ACCEPT
```
This allows web traffic to your system if you run a web server[3][8].

---

### **4. Block All Other Incoming Traffic**

With the default INPUT policy set to DROP, all other incoming traffic is automatically blocked unless you add more rules[8].

---

````python
import os
import hashlib
import shutil
import random
import time
import datetime

TEST_DIR = "virus_lab_environment"

def setup_test_environment():
    if os.path.exists(TEST_DIR):
        shutil.rmtree(TEST_DIR)
    os.makedirs(TEST_DIR)
    for i in range(5):
        with open(f"{TEST_DIR}/normal_file_{i}.txt", "w") as f:
            f.write(f"This is normal file {i} content.\n")
            f.write("It contains legitimate information.\n")

class SimulatedVirus:
    def __init__(self):
        self.signature = "# BEGIN-HARMLESS-VIRUS-SIMULATION\n"
        self.end_signature = "# END-HARMLESS-VIRUS-SIMULATION\n"
        self.virus_code = """
def replicate():
    import os
    files_in_current_dir = os.listdir('.')
    for filename in files_in_current_dir:
        if filename.endswith('.txt') and "infected" not in filename:
            try:
                with open(filename, 'a') as file:
                    file.write("# File has been simulated-infected\\n")
            except:
                pass
    return "Simulation complete"

print("This is a simulated virus payload!")
replicate()
"""
        self.infected_files = []

    def infect_file(self, file_path):
        try:
            with open(file_path, 'r') as f:
                content = f.read()
                if self.signature in content:
                    return False
            with open(file_path, 'a') as f:
                f.write("\n" + self.signature)
                f.write(self.virus_code)
                f.write(self.end_signature)
            self.infected_files.append(file_path)
            return True
        except Exception as e:
            print(f"Infection failed for {file_path}: {str(e)}")
            return False

    def spread(self, directory):
        self.infected_files = []
        for filename in os.listdir(directory):
            file_path = os.path.join(directory, filename)
            if os.path.isfile(file_path) and filename.endswith('.txt'):
                self.infect_file(file_path)
        return len(self.infected_files), self.infected_files

class SimpleAntivirus:
    def __init__(self):
        self.known_signatures = [
            "# BEGIN-HARMLESS-VIRUS-SIMULATION",
            "# File has been simulated-infected"
        ]
        self.clean_file_hashes = {}
        self.detected_files = []
        self.cleaned_files = []

    def scan_file(self, file_path):
        try:
            with open(file_path, 'r') as f:
                content = f.read()
                for signature in self.known_signatures:
                    if signature in content:
                        if file_path not in self.detected_files:
                            self.detected_files.append(file_path)
                        return True, signature
                return False, None
        except Exception as e:
            return False, f"Error reading file: {str(e)}"

    def clean_file(self, file_path):
        try:
            with open(file_path, 'r') as f:
                content = f.read()
            start_marker = "# BEGIN-HARMLESS-VIRUS-SIMULATION"
            end_marker = "# END-HARMLESS-VIRUS-SIMULATION"

            start_index = content.find(start_marker)
            if start_index == -1:
                return False, "Virus signature not found"

            end_index = content.find(end_marker, start_index)
            if end_index == -1:
                end_index = len(content)
            else:
                end_index += len(end_marker)

            cleaned_content = content[:start_index] + content[end_index:]
            with open(file_path, 'w') as f:
                f.write(cleaned_content)
            if file_path not in self.cleaned_files:
                self.cleaned_files.append(file_path)
            return True, "File cleaned successfully"
        except Exception as e:
            return False, f"Error cleaning file: {str(e)}"

    def update_definitions(self, new_signature):
        if new_signature not in self.known_signatures:
            self.known_signatures.append(new_signature)

    def build_clean_file_database(self, directory):
        for filename in os.listdir(directory):
            file_path = os.path.join(directory, filename)
            if os.path.isfile(file_path):
                with open(file_path, 'rb') as f:
                    file_hash = hashlib.md5(f.read()).hexdigest()
                self.clean_file_hashes[file_path] = file_hash

    def scan_directory(self, directory):
        self.detected_files = []
        self.cleaned_files = []

        results = {
            'scanned': 0,
            'infected': 0,
            'cleaned': 0,
            'infected_files': [],
            'cleaned_files': []
        }

        for filename in os.listdir(directory):
            file_path = os.path.join(directory, filename)
            if os.path.isfile(file_path):
                results['scanned'] += 1
                infected, signature = self.scan_file(file_path)

                if infected:
                    results['infected'] += 1
                    results['infected_files'].append(file_path)

                    cleaned, message = self.clean_file(file_path)
                    if cleaned:
                        results['cleaned'] += 1
                        results['cleaned_files'].append(file_path)

        return results

    def generate_report(self, output_file="antivirus_report.txt"):
        report = f"""
=======================================================
ANTIVIRUS SCAN REPORT
=======================================================
Date and Time: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

SUMMARY:
-------------------------------------------------------
Files Scanned: {len(self.detected_files) + len(self.clean_file_hashes)}
Infected Files Found: {len(self.detected_files)}
Files Cleaned: {len(self.cleaned_files)}

DETECTED INFECTIONS:
-------------------------------------------------------
"""
        for file in self.detected_files:
            report += f"- {file}\n"

        report += """
CLEANING ACTIONS:
-------------------------------------------------------
"""
        for file in self.cleaned_files:
            report += f"- {file}\n"

        report += """
=======================================================
End of Report
=======================================================
"""
        with open(output_file, "w") as f:
            f.write(report)

        return report

def demonstrate():
    setup_test_environment()
    print(f"Created test environment in '{TEST_DIR}' directory")
    print(f"Created these files:")
    for filename in os.listdir(TEST_DIR):
        print(f"  - {os.path.join(TEST_DIR, filename)}")

    virus = SimulatedVirus()
    antivirus = SimpleAntivirus()

    print("\n=== INITIAL SCAN ===")
    scan_results = antivirus.scan_directory(TEST_DIR)
    print(f"Scanned {scan_results['scanned']} files")
    print(f"Found {scan_results['infected']} infected files")

    print("\n=== VIRUS ACTIVITY ===")
    infected_count, infected_files = virus.spread(TEST_DIR)
    print(f"Virus infected {infected_count} files:")
    for file in infected_files:
        print(f"  - {file}")

    print("\n=== INFECTION DETECTION ===")
    scan_results = antivirus.scan_directory(TEST_DIR)
    print(f"Scanned {scan_results['scanned']} files")
    print(f"Found {scan_results['infected']} infected files:")
    for file in scan_results['infected_files']:
        print(f"  - {file}")

    print(f"\nCleaned {scan_results['cleaned']} files:")
    for file in scan_results['cleaned_files']:
        print(f"  - {file}")

    print("\n=== VERIFICATION SCAN ===")
    scan_results = antivirus.scan_directory(TEST_DIR)
    print(f"Scanned {scan_results['scanned']} files")
    print(f"Found {scan_results['infected']} infected files")

    report = antivirus.generate_report(f"{TEST_DIR}/antivirus_report.txt")
    print(f"\nDetailed report generated at: {TEST_DIR}/antivirus_report.txt")

if __name__ == "__main__":
    print("===== VIRUS AND ANTIVIRUS SIMULATION FOR EDUCATIONAL PURPOSES =====")
    print("This script demonstrates basic virus and antivirus concepts safely.")
    demonstrate()
````
